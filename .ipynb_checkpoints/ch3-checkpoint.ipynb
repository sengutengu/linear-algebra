# Chapter 3. Invertibility, bases, and coordinate systems

## 3.1 Invertibility

### 3.1.1 Invertible matrices

A square matrix $A$ is invertible if there is a matrix $B$ such that $AB = I$, where $I$ is the identity matrix of the same square dimensions. $B$ is called the **inverse** of $A$ and is denoted $A^{-1}$.

#### **Activity 3.1.2**

A. Suppose that $A = \begin{bmatrix}
3 & -2\\
1 & -1
\end{bmatrix}$. To find the inverse $B$, we require that $AB = I$, and more specifically, 

$$\begin{bmatrix}
Ab_1 & Ab_2
\end{bmatrix} = \begin{bmatrix}
1 & 0\\
0 & 1
\end{bmatrix}$$

So, we can solve for $B$ by finding $Ab_1 = \begin{bmatrix}
1 \\
0
\end{bmatrix}$ and $Ab_2 = \begin{bmatrix}
0 \\
1
\end{bmatrix}$.

```python
from sympy import Matrix
```

```python
A = Matrix(2,2,[3,-2,1,-1])
b1 = Matrix(2,1,[1,0])
b2 = Matrix(2,1,[0,1])

print(A.row_join(b1).rref())
print(A.row_join(b2).rref())
```

So, $B = \begin{bmatrix}
1 & -2\\
1 & -3
\end{bmatrix}$.

B. The product $BA$ is, as we would expect:

$BA = \begin{bmatrix}
1 & -2\\
1 & -3
\end{bmatrix}\begin{bmatrix}
3 & -2\\
1 & -1
\end{bmatrix} = \begin{bmatrix}
1 & 0\\
0 & 1
\end{bmatrix}$. We have found the inverse of $A$. Since $B$ is the inverse of $A$, $B$ is also invertible, and $B^{-1} = A$.

C. Let's now try to find the inverse of $C = \begin{bmatrix}
-2 & 1\\
4 & -2
\end{bmatrix}$. When we repeat the above procedure, we find that the systems are inconsistent.

```python
C = Matrix(2,2,[-2,1,4,-2])

print(C.row_join(b1).rref())
print(C.row_join(b2).rref())
```

D. If $A$ is an invertible $n\times n$ matrix with the inverse $B$, and $b$ is **any** $n$-dimensional vector, because $AB = I$, we obtain: 

$$A(Bb) = (AB)b = Ib = b.$$

And this means that $Ax = b$ is consistent, with $x = Bb$ being a solution. Just to be clear, this means that there exists some $x$ that satisfies $Ax = b$ (i.e. there is some $x$ that, when fed into $T(x)$, returns the output $b$), and that $x$ is equal to $Bb$, which is the inverse of $A$ times $b$. 

What does this say about the span of the columns of $A$? **Solution**. The span of the columns of $A$ is $\mathbb{R}^n$. Every vector in $\mathbb{R}^n$ can be obtained by some linear combination of vectors in $A$, because every vector can be multiplied by $B$ and then fed into $A$ to yield itself. 

E. Since $A$ is a square matrix, what does this say about the pivot positiosn of $A$? What is the RREF of $A$? **Solution**. Every column must contain a pivot position. The RREF is just the identity matrix $I$. In contrast, the matrix $C$ has the RREF 

$$\begin{bmatrix}
1 & -1/2\\
0 & 0
\end{bmatrix}$$

So invertible matrices must have a pivot in every row and every column, meaning their RREFs must be the identity matrix.

We can reformulate the procedure for finding the inverse of a matrix. Rather than finding $Ab_1$ and $Ab_2$ separately, we can just augment $A$ by the identity matrix and find the RREF. 

For example, given $A=\begin{bmatrix}
1 & 2\\
1 & 1
\end{bmatrix}$, we create the augmented matrix:

$$
\left[\begin{array}{cc|cc}
1 & 2 & 1 & 0  \\
1 & 1 & 0 & 1
\end{array}\right] = \left[\begin{array}{cc|cc}
1 & 0 & -1 & 2  \\
0 & 1 & 1 & -1
\end{array}\right]$$

### 3.1.2 Solving equations with an inverse

If $A$ is an invertible matrix, then the solution to $Ax = b$ is $x = A^{-1}b$.

#### **Activity 3.1.3**

Consider the square matrix $A = \begin{bmatrix}
1 & 0 & 2\\
2 & 2 & 1 \\
1 & 1 & 1
\end{bmatrix}.$

A. Describe the solution space to the equation $Ax = \begin{bmatrix}
3 \\
4 \\
3
\end{bmatrix}$. **Solution**. $x = \begin{bmatrix}
-1 \\
2 \\
2
\end{bmatrix}$. The solution space is the single point $x$.

```python
A = Matrix(3,3,[1,0,2,2,2,1,1,1,1])
b = Matrix(3,1,[3,4,3])

A.row_join(b).rref()
```

B. Why is $A$ invertible? Find the inverse. **Solution**. $A$ is invertible because it is consistent and its RREF is the identity matrix. The inverse $A^{-1} = \begin{bmatrix}
1 & 2 & -4\\
-1 & -1 & 3 \\
0 & -1 & 2
\end{bmatrix}$.

```python
A.row_join(Matrix(3,3,[1,0,0,0,1,0,0,0,1])).rref()
```

C. Now, use the inverse to solve the equation $Ax = \begin{bmatrix}
3 \\
4 \\
3
\end{bmatrix}$ and verify that your result agrees with what you found in A. **Solution**. 

$$A^{-1}\begin{bmatrix}
3 \\
4 \\
3
\end{bmatrix} = \begin{bmatrix}
-1 \\
2 \\
2
\end{bmatrix}.$$

```python
A.inv()*Matrix(3,1,[3,4,3])
```

D. Find the inverse of the matrix

$$B = \begin{bmatrix}
1 & -2 & -1\\
-1 & 5 & 6 \\
5 & -4 & 6
\end{bmatrix}$$

and use it to solve the equation $Bx = \begin{bmatrix}8\\3\\36\end{bmatrix}$.

**Solution**. $x = \begin{bmatrix}76\\47\\-26\end{bmatrix}$.

```python
B = Matrix(3,3,[1,-2,-1,-1,5,6,5,-4,6])
B.inv()*Matrix(3,1,[8,3,36])
```

E. Find the product $AB$ and verify that it is invertible.

**Solution**. Computing the RREF of the product $AB$ returns the identity matrix, so it is invertible.

```python
A*B
```

```python
(A*B).rref()
```

F. Compute the products $A^{-1}B^{-1}$ and $B^{-1}A^{-1}$. Which one agrees with $(AB)^{-1}$?

**Solution**. $B^{-1}A^{-1}$ agrees with $(AB)^{-1}$.

```python
(A*B).inv()
```

```python
A.inv()*B.inv()
```

```python
B.inv()*A.inv()
```

G. Explain your finding by considering the product $(AB)(B^{-1}A^{-1})$ and using associativity to regroup the products so that the middle two terms are multiplied first.

**Solution**. If $(B^{-1}A^{-1})$ equals $(AB)^{-1}$, i.e. the inverse of $AB$, then we'd expect $(AB)(B^{-1}A^{-1}) = I$. Multiplying the middle two terms, we obtain $AIA^{-1} = AA^{-1} = I$. Therefore, $(B^{-1}A^{-1}) = (AB)^{-1}$.

### Interim Summary
* A square matrix is invertible if and only if $A \sim I$.
* If $A$ is invertible, then the solution to $Ax = b$ is $x = A^{-1}b$.
* We can find $A^{-1}$ by finding the RREF of $[A | I]$.
* If $A$ and $B$ are two invertible matrices of the same size, then $AB$ is also invertible, and $(AB)^{-1} = B^{-1}A^{-1}$.
* A $2 \times 2$ matrix is invertible if the determinant $(ad-bc) \neq 0$. We'll talk about determinants later.
* The inverse of a $2 \times 2$ matrix is $\frac{1}{\text{det}}\begin{bmatrix}d & -b \\ -c & a\end{bmatrix}$.

### 3.1.3 Triangular matrices and Gaussian elimination

A matrix is **lower triangular** if all entries above the diagonal are zero; it is **upper triangular** if all entries below the diagonal are zero.

For example, $L$ is a lower triangular matrix and $U$ is an upper triangular matrix.

$$L = \begin{bmatrix}* & 0 & 0 \\ * & * & 0 \\ * & * & *\end{bmatrix}, U = \begin{bmatrix}* & * & * \\ 0 & * & * \\ 0 & 0 & *\end{bmatrix}$$

Given a lower triangular matrix, if we take the RREF and find that all entries on the diagonal are nonzero, then we know that the matrix is invertible.

#### **Activity 3.1.4**

Consider the matrix

$$A = \begin{bmatrix}1 & 2 & 1 \\ 2 & 0 & -2 \\ -1 & 2 & -1\end{bmatrix}.$$

To begin Gaussian elimination, we multiply Row 1 by $-2$ and add it to Row 2. But instead of applying the operation directly to $A$, we perform the operation onto $I$. So,

$$I = \begin{bmatrix}1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1\end{bmatrix} \to R = \begin{bmatrix}1 & 0 & 0 \\ -2 & 1 & 0 \\ 0 & 0 & 1\end{bmatrix}$$ 

**A. Show that the product $RA$ is the result of applying the replacement operation to $A$.**

**Solution**. 

```python
A = Matrix(3,3,[1,2,1,2,0,-2,-1,2,-1])
R = Matrix(3,3,[1,0,0,-2,1,0,0,0,1])
R*A
```

**B. Explain why $R$ is invertible and find the inverse.**

**Solution**. $R$ is invertible because (in addition to reducing to $I$), it is a lower triangular matrix with all nonzero entries on the diagonal. 

```python
R.inv()
```

**C. Describe the relationship between $R$ and $R^{-1}$ and use the connection to replacement operations to explain why it holds.**

**Solution**. It looks like, whereas $R$ multiplies Row 1 by $-2$ and adds it to Row 2, $R^{-1}$ multiplies Row 1 by $2$ and adds it to Row 2. So, $R^{-1}$ is undoing (inverting) the addition operation performed by $R$.

**D. Other row operations can be performed using a similar procedure. For instance, suppose we want to scale the second row of $A$ by $4$. Find a matrix $S$ so that $SA$ is the same as that obtained from the scaling operation. Why is $S$ invertible, and what is $S^{-1}$?**

**Solution**. 

$$S = \begin{bmatrix}1 & 0 & 0 \\ 0 & 4 & 0 \\ 0 & 0 & 1\end{bmatrix}.$$ 

$S$ is invertible because it is a lower triangle matrix with all nonzero entries on the diagonal, and $S^{-1} = \begin{bmatrix}1 & 0 & 0 \\ 0 & \frac{1}{4} & 0 \\ 0 & 0 & 1\end{bmatrix}$

```python
S = Matrix(3,3,[1,0,0,0,4,0,0,0,1])
S*A
S.inv()
```

**E. Finally, we want to interchange the first and third rows of $A$. Find matrix $P$, usually called a *permutation matrix*. What is $P^{-1}$?**

**Solution**. 

$$P = \begin{bmatrix}0 & 0 & 1 \\ 0 & 1 & 0 \\ 1 & 0 & 0\end{bmatrix}, P^{-1} = \begin{bmatrix}0 & 0 & 1 \\ 0 & 1 & 0 \\ 1 & 0 & 0\end{bmatrix}.$$

$P^{-1}$ feels a bit surprising, but upon reflection, it makes perfect sense: to undo the interchange between Row 1 and Row 3, we just interchange them again.

```python
P = Matrix(3,3,[0,0,1,0,1,0,1,0,0])
P*A
P.inv()
```

**F. The original matrix $A$ is row equivalent to the upper triangular matrix $U$ by performing three replacement operations on $A$:**

$$A = \begin{bmatrix}1 & 2 & 1 \\ 2 & 0 & -2 \\-1 & 2 & -1\end{bmatrix} \sim \begin{bmatrix}1 & 2 & 1 \\ 0 & -4 & -4 \\0 & 0 & -4\end{bmatrix} = U.$$

Find matrices $L_1$, $L_2$, and $L_3$ that perform these row replacement operations so that $L_3L_2L_1A=U$.

**Solution**.

$$L_1 = \begin{bmatrix}1 & 0 & 0 \\ -2 & 1 & 0 \\ 0 & 0 & 1\end{bmatrix}, L_2 = \begin{bmatrix}1 & 0 & 0 \\ 0 & 1 & 0 \\ 1 & 0 & 1\end{bmatrix}, L_3 = \begin{bmatrix}1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 1 & 1\end{bmatrix}$$

```python
L1 = Matrix(3,3,[1,0,0,-2,1,0,0,0,1]) # Multiply Row 1 by 2 then subtract from Row 2
L2 = Matrix(3,3,[1,0,0,0,1,0,1,0,1]) # Add Row 1 to Row 2
L3 = Matrix(3,3,[1,0,0,0,1,0,0,1,1]) # Add Row 2 to Row 3
L3*L2*L1*A
```

**G. Explain why $L_3L_2L_1$ is invertible, and use this fact to write $A = LU$. What is the matrix $L$? Why do we denote it by $L$?**

**Solution**. $L_3L_2L_1$ is invertible because it is a triangular matrix with all nonzero values on the matrix. This means that $L_3L_2L_1$ has an inverse, and we can find $A = LU$, if we treat $L_3L_2L_1 = L^{-1}$. Then, $L$ is as below:

```python
L1 = Matrix(3,3,[1,0,0,-2,1,0,0,0,1]) # Multiply Row 1 by 2 then subtract from Row 2
L2 = Matrix(3,3,[1,0,0,0,1,0,1,0,1]) # Add Row 1 to Row 2
L3 = Matrix(3,3,[1,0,0,0,1,0,0,1,1]) # Add Row 2 to Row 3
L = (L3*L2*L1).inv()
L
```

Then, $A = LU = \begin{bmatrix}1 & 0 & 0 \\ 2 & 1 & 0 \\ -1 & -1 & 1\end{bmatrix}\begin{bmatrix}1 & 2 & 1 \\ 0 & -4 & -4 \\0 & 0 & -4\end{bmatrix}$.

```python
U = Matrix(3,3,[1,2,1,0,-4,-4,0,0,-4])
A = L*U
A
```
