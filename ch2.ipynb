## 2.5 Matrix transformations

A **function** maps each element of one set (the **domain**) to exactly one element of another set (the **codomain**). Often, we think of a function as a machine, $f$, that performs a job—a *function*—on some input, $x$, and spits out exactly one output, $f(x)$. 



Graphically, a function can be conceived of as a **transformation** of a straight $y=x$ line. For example, a function $g(x) = 2x$ transforms $y=x$ by rotating it about the origin +15 degrees; a function $h(x) = x^2$ transforms $y=x$ by multiplying $x$ by itself, bending the line up. 

There are, of course, functions in linear algebra, and matrices encode them. The **matrix transformation** associated with the matrix $A$ is the function $T(x)$ that assigns the vector $Ax$ to the vector $x$. Let's take a look at an example:

### Example 2.5.4

The matrix $A = \begin{bmatrix} 
3 & -2 \\
1 & 2 \\
0 & 3 
\end{bmatrix}
$ defines a matrix transformation $T(x) = Ax$ such that 

$$T\left(\begin{bmatrix}
x \\
y
\end{bmatrix}\right) = \begin{bmatrix} 
3 & -2 \\
1 & 2 \\
0 & 3 
\end{bmatrix}\begin{bmatrix}
x \\
y
\end{bmatrix} = \begin{bmatrix}
3x - 2y \\
x + 2y \\
3y
\end{bmatrix}$$. 

Since inputs to the function $T$ must be two-dimensional, and the output is three-dimensional, we write $T: \mathbb{R}^2 \to \mathbb{R}^3$. Importantly, we see that **the matrix $A$ stores the coefficients of linear transformations performed on the input vector $\begin{bmatrix}
x \\
y
\end{bmatrix}$**.

### Example 2.5.5

If we have a function $T: \mathbb{R}^3 \to \mathbb{R}^2$ that has the form $T\left(\begin{bmatrix}
x \\
y \\
z
\end{bmatrix}\right) = \begin{bmatrix} 
-4x -y + 2z \\
x + 2y -z
\end{bmatrix}$, we can figure out what matrix $A$ defines this function. Since the matrix $A$ stores the coefficients of the linear transformations on the input vector $x$, we just take the coefficients in the output matrix and stick them into a matrix:

$$A = \begin{bmatrix} 
-4 & -1 & 2 \\
1 & 2 & -1
\end{bmatrix}$$

### Composing matrix transformations
Just as functions can be composed, matrix transformations can also be composed. If $T$ is associated with matrix $A$, and $S$ is associated with matrix $B$, then $(S \circ T)$ is associated with $BA$.

### Activity 2.5.4

A. Let's try composing the $90\degree$-rotation matrix transformation by itself.
1. What is the matrix associated with the composition $(T\circ T)$? **Solution**. Recall that $A = \begin{bmatrix} 
0 & -1\\
1 & 0
\end{bmatrix}$. So, $AA = \begin{bmatrix} 
0 & -1\\
1 & 0
\end{bmatrix}\begin{bmatrix} 
0 & -1\\
1 & 0
\end{bmatrix} = \begin{bmatrix} 
-1 & 0\\
0 & -1
\end{bmatrix}$. This is consistent with the fact that when we rotate $90\degree$ twice, we rotate by $180\degree$.

B. In the previous activity, we determined the relationship between bakery hours and number of good produced. We can compose the previous relationship with the relationship between the type of goods and ingredients required. Suppose now that

* Each cake requires 4 cups of flour and and 2 cups of sugar.
* Each doughnut requires 1 cup of flour and 1 cup of sugar.
* Each eclair requires 1 cup of flour and 2 cups of sugar.

Matrix transformation $S: \mathbb{R}^3 \to \mathbb{R}^2$ takes the number of cakes, doughnuts, and eclairs as input and produces the amount of flour and sugar as output. 

1. Write the matrix $B$ associated with transformation $S$. **Solution**. $B = \begin{bmatrix} 
4 & 1 & 1\\
2 & 1 & 2
\end{bmatrix}$.
2. If we make 1200 cakes, 2850 doughnuts, and 2250 eclairs, how many cups of flour and sugar are required? **Solution**. $\begin{bmatrix}

```python
B = Matrix(2,3,[4,1,1,2,1,2])
x = Matrix(3,1,[1200,2850,2250])
B*x
```

3. If Bakery 1 operates for 75 hours and Bakery 2 operates for 53 hours, how many cakes, doughnuts, and eclairs are produced, and how many cups of flour and sugar are required? **Solution**. 1810 cakes, 5340 doughnuts, and 3840 eclairs; 16420 cups of flour and 16640 cups of sugar.

```python
A = Matrix(3,2,[10,20,50,30,30,30])
x = Matrix(2,1,[75, 53])
B = Matrix(2,3,[4,1,1,2,1,2])
print(A*x)
print(B*A*x)
```
