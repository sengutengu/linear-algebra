## 3.1 Invertibility

A square matrix $A$ is invertible if there is a matrix $B$ such that $AB = I$, where $I$ is the identity matrix of the same square dimensions. $B$ is called the **inverse** of $A$ and is denoted $A^{-1}$.

### Activity 3.1.2

A. Suppose that $A = \begin{bmatrix}
3 & -2\\
1 & -1
\end{bmatrix}$. To find the inverse $B$, we require that $AB = I$, and more specifically, 

$$\begin{bmatrix}
Ab_1 & Ab_2
\end{bmatrix} = \begin{bmatrix}
1 & 0\\
0 & 1
\end{bmatrix}$$

So, we can solve for $B$ by finding $Ab_1 = \begin{bmatrix}
1 \\
0
\end{bmatrix}$ and $Ab_2 = \begin{bmatrix}
0 \\
1
\end{bmatrix}$.

```python
from sympy import Matrix
```

```python
A = Matrix(2,2,[3,-2,1,-1])
b1 = Matrix(2,1,[1,0])
b2 = Matrix(2,1,[0,1])

print(A.row_join(b1).rref())
print(A.row_join(b2).rref())
```

So, $B = \begin{bmatrix}
1 & -2\\
1 & -3
\end{bmatrix}$.

B. The product $BA$ is, as we would expect:

$BA = \begin{bmatrix}
1 & -2\\
1 & -3
\end{bmatrix}\begin{bmatrix}
3 & -2\\
1 & -1
\end{bmatrix} = \begin{bmatrix}
1 & 0\\
0 & 1
\end{bmatrix}$. We have found the inverse of $A$. Since $B$ is the inverse of $A$, $B$ is also invertible, and $B^{-1} = A$.

C. Let's now try to find the inverse of $C = \begin{bmatrix}
-2 & 1\\
4 & -2
\end{bmatrix}$. When we repeat the above procedure, we find that the systems are inconsistent.

```python
C = Matrix(2,2,[-2,1,4,-2])

print(C.row_join(b1).rref())
print(C.row_join(b2).rref())
```

D. If $A$ is an invertible $n\times n$ matrix with the inverse $B$, and $b$ is **any** $n$-dimensional vector, because $AB = I$, we obtain: 

$$A(Bb) = (AB)b = Ib = b.$$

And this means that $Ax = b$ is consistent, with $x = Bb$ being a solution. Just to be clear, this means that there exists some $x$ that satisfies $Ax = b$ (i.e. there is some $x$ that, when fed into $T(x)$, returns the output $b$), and that $x$ is equal to $Bb$, which is the inverse of $A$ times $b$. 

What does this say about the span of the columns of $A$? **Solution**. The span of the columns of $A$ is $\mathbb{R}^n$. Every vector in $\mathbb{R}^n$ can be obtained by some linear combination of vectors in $A$, because every vector can be multiplied by $B$ and then fed into $A$ to yield itself. 

E. Since $A$ is a square matrix, what does this say about the pivot positiosn of $A$? What is the RREF of $A$? **Solution**. Every column must contain a pivot position. The RREF is just the identity matrix $I$. In contrast, the matrix $C$ has the RREF 

$$\begin{bmatrix}
1 & -1/2\\
0 & 0
\end{bmatrix}$$

So invertible matrices must have a pivot in every row and every column, meaning their RREFs must be the identity matrix.

We can reformulate the procedure for finding the inverse of a matrix. Rather than finding $Ab_1$ and $Ab_2$ separately, we can just augment $A$ by the identity matrix and find the RREF. 

For example, given $A=\begin{bmatrix}
1 & 2\\
1 & 1
\end{bmatrix}$, we create the augmented matrix:

$$
\left[\begin{array}{cc|cc}
1 & 2 & 1 & 0  \\
1 & 1 & 0 & 1
\end{array}\right] = \left[\begin{array}{cc|cc}
1 & 0 & -1 & 2  \\
0 & 1 & 1 & -1
\end{array}\right]$$
